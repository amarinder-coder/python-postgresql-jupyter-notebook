Calling to create a PostgreSql database.

```python
cur, conn = create_database()
```

```python
country_t_c = ("""CREATE TABLE IF NOT EXISTS accountscountry(
country_code VARCHAR PRIMARY KEY,
short_name VARCHAR,
table_name VARCHAR,
long_name VARCHAR,
currency_unit VARCHAR
)""")
```

```python
cur.execute(country_t_c)
conn.commit()
```

```python
accounts_t_c = ("""CREATE TABLE IF NOT EXISTS accountsdata(
country_name VARCHAR,
country_code VARCHAR,
indicator_name VARCHAR,
indicator_code VARCHAR,
year_2000 numeric,
year_2005 numeric,
year_2010 numeric,
year_2014 numeric
)""")
```

```python
cur.execute(accounts_t_c)
conn.commit()
```

```python
series_t_c = ("""CREATE TABLE IF NOT EXISTS accountseries(
series_code VARCHAR,
topic VARCHAR,
short_definition VARCHAR,
indicator_code VARCHAR
)""")
```

```python
cur.execute(series_t_c)
conn.commit()
```
